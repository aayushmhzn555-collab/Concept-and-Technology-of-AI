```python
# Task 1 
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            total += sum_nested_list(element)
        else:
            total += element
    return total

# Test:
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
print("sum_nested_list test:", sum_nested_list(nested_list)) 
```

```python
# Task 2 
def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    perms = []
    seen = set()
    for i, ch in enumerate(s):
        if (ch, s[:i], s[i+1:]) in seen:
            continue
        remaining = s[:i] + s[i+1:]
        for p in generate_permutations(remaining):
            perms.append(ch + p)
        seen.add((ch, s[:i], s[i+1:]))
    return list(dict.fromkeys(perms))

print("perms 'abc':", sorted(generate_permutations("abc")))
print("perms 'aab':", sorted(generate_permutations("aab")))
```

```python
# Task 3 
def calculate_directory_size(directory):
    total = 0
    for name, val in directory.items():
        if isinstance(val, dict):
            total += calculate_directory_size(val)
        else:
            total += val
    return total

directory_structure = {
 "file1.txt": 200,
 "file2.txt": 300,
 "subdir1": {
  "file3.txt": 400,
  "file4.txt": 100
 },
 "subdir2": {
  "subsubdir1": {
   "file5.txt": 250
  },
  "file6.txt": 150
 }
}

print("Directory size (KB):", calculate_directory_size(directory_structure)) 
```

```python
# Task 1 
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float('inf') else -1

print("min_coins([1,2,5], 11) =", min_coins([1,2,5], 11))  
print("min_coins([2], 3) =", min_coins([2], 3))        
```

```python
# Task 2 
def longest_common_subsequence(s1, s2):
    n, m = len(s1), len(s2)
    dp = [[0] * (m + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            if s1[i-1] == s2[j-1]:
                dp[i][j] = dp[i-1][j-1] + 1
            else:
                dp[i][j] = max(dp[i-1][j], dp[i][j-1])
    return dp[n][m]

print("LCS('abcde','ace') =", longest_common_subsequence("abcde","ace")) 
```

```python
# Task 3 
def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]
    for i in range(1, n + 1):
        wt = weights[i-1]
        val = values[i-1]
        for w in range(capacity + 1):
            dp[i][w] = dp[i-1][w]  
            if wt <= w:
                dp[i][w] = max(dp[i][w], dp[i-1][w-wt] + val)
    return dp[n][capacity]

print("knapsack test:", knapsack([1,3,4,5],[1,4,5,7],7))
```
